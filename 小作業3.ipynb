# 在學校時，常常遇到一個惱人的問題。

         就是不知道*等等午餐吃什麼*。

         所以請設計一個程式，運用while、if，替大家解決此問題。

         (while用來增加程式的在使用性，if做為選項之判斷)

     

         條件如下:1.請確認中式或西式(如:想吃中式就輸入1，想吃西式就輸入2)
                            如果輸入不是數字，則顯示"輸入格式錯誤"
                          2.請確認預算大小，以50以下、50~80，81~100為分界(預算50以下請打1，50~80就打2，81~100就打3)
                             如果輸入不是數字，則顯示"輸入格式錯誤"
                          3.再依據上方兩個條件做配對，共6組

                          4.請預先設置總共6組餐點選擇答案

```python
a=input('輸入中式50元以下的餐點:')
b=input('輸入中式50~80元的餐點:')
c=input('輸入中式81~100元的餐點:')

A=input('輸入西式50元以下的餐點:')
B=input('輸入西式50~80元的餐點:')
C=input('輸入西式81~100元的餐點:')

i=0
i=int(i)

while i>=0:
    style=input('請輸入想吃中式(1)還是西式(2)')
    if style.isdecimal():
        style=int(style)
        if  2>style>0:
            print('還敢吃中式啊')
            while i >-1:
                bud=input('預算大小?預算50以下打1，50~80打2，81~100打3')
                if bud.isdecimal():
                    bud=int(bud)
                    if bud==1:
                        print(a)
                        i=-1
                    elif bud==2:
                        print(b)
                        i=-1
                    elif bud==3:
                        print(c)
                        i=-1
                    else:
                        print('輸入格式錯誤')
                else:
                    print('輸入格式錯誤')
        elif style==2:
            print('Tryna eat western my BOI?')
            while i>-1:
                bud=input('Your budget?Plz type 1 if budget is below 50，2 if 50-80，3 if 81-100')
                if bud.isdecimal():
                    bud=int(bud)
                    if bud==1:
                        print(A)
                        i=-1
                    elif bud==2:
                        print(B)
                        i=-1
                    elif bud==3:
                        print(C)
                        i=-1
                    else:
                        print('輸入格式錯誤')
                else:
                    print('輸入格式錯誤')
        else:
            print('輸入格式錯誤') 
    else:
        print('輸入格式錯誤')
```
