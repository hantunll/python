```python
def avg(a):#定義算每日平均的函數
    aa=a/7
    aaa=round(aa)
    return aaa
def ratio(a,A):#定義算百分比的函數
    return round(100*a/A)
day=[]#空串列，用來放超過日支出的串列
list1=[]#空串列，用來第1項支出的串列
list2=[]#空串列，用來第2項支出的串列
list3=[]#空串列，用來第3項支出的串列
list4=[]#空串列，用來第4項支出的串列
count=0#計數器

#先選擇功能
print('歡迎來到記帳世界')
print('功能:')
print('1.依天記錄')
print('2.分類紀錄')
mode=input('請輸入想要的功能(輸入代碼1或2)(要離開請輸入0)')

while mode!='0':    
    while mode.isdecimal()==True:
        mode1=int(mode)
#功能一:依天記錄
        if mode1==1:
            d=input('請輸入每日預算')
            i = 0
            exceed_day=[]
            day_cost_list=[]
            while not d.isdecimal() :
                d=input('請輸入每日預算')
            dd=int(d)
            day_cost='x'
            while (not day_cost.isdecimal()) or i<7:
                day_cost=input('請輸入每日支出')
                if day_cost.isdecimal():
                    day_cost_list.append(int(day_cost))
                    if int(day_cost)>dd:
                        exceed_day.append(i)
                    i=i+1
            if len(exceed_day)!=0:
                print('本周超出預算的日子')
                for item in exceed_day:
                    print('星期{0:d}'.format(item+1))
            else:#超出的串列中沒有元素
                print('本周每日皆無超出預算')
            total=sum(day_cost_list) #總支出
            print('-'*30)#分隔線
            print('本周總支出為'+str(total)+'元')
            print('本每日平均支出為'+str(avg(total))+'元')
            print('-'*30)
            print('本周的支出比例')#每日比例的結果分析
            print('{0:<s}{1:^10s}{2:>s}' .format('日期','支出','比例'))#日期:靠右；支出:置中；比例:靠右
            print('-'*30)
            print('{0:<s}{1:^10d}{2:>d}' .format('星期一',int(day_cost_list[0]), ratio(int(day_cost_list[0]),total))+'%')
            print('{0:<s}{1:^10d}{2:>d}' .format('星期二',int(day_cost_list[1]), ratio(int(day_cost_list[1]),total))+'%')
            print('{0:<s}{1:^10d}{2:>d}' .format('星期三',int(day_cost_list[2]), ratio(int(day_cost_list[2]),total))+'%')
            print('{0:<s}{1:^10d}{2:>d}' .format('星期四',int(day_cost_list[3]), ratio(int(day_cost_list[3]),total))+'%')
            print('{0:<s}{1:^10d}{2:>d}' .format('星期五',int(day_cost_list[4]), ratio(int(day_cost_list[4]),total))+'%')
            print('{0:<s}{1:^10d}{2:>d}' .format('星期六',int(day_cost_list[5]), ratio(int(day_cost_list[5]),total))+'%')
            print('{0:<s}{1:^10d}{2:>d}' .format('星期日',int(day_cost_list[6]), ratio(int(day_cost_list[6]),total))+'%')
            mode=input('請輸入想要的功能(輸入代碼1或2)(要離開請輸入0)')
            if mode=='0':
                break



                                           
        #功能二:依項目紀錄
        elif mode1==2 :
            y1=input('請輸入飲食支出預算')
            while y1.isdecimal()==False:#判斷預算輸入是否有誤
                print('error')
                y1=input('請輸入飲食支出預算')
            yy1=int(y1)
            y2=input('請輸入日用品支出預算')
            while  y2.isdecimal()==False:#判斷預算輸入是否有誤
                print('error')
                y2=input('請輸入日用品支出預算')
            yy2=int(y2)
            y3=input('請輸入娛樂支出預算')
            while  y3.isdecimal()==False:#判斷預算輸入是否有誤
                print('error')
                y3=input('請輸入娛樂支出預算')
            yy3=int(y3) 
            y4=input('請輸入其他預算支出')
            while y4.isdecimal()==False:#判斷預算輸入是否有誤
                print('error')
                y4=input('請輸入其他預算支出')
            yy4=int(y4)
            print('1.飲食 2.日用品 3.娛樂 4.其他 0:離開')
            types=input('請選擇要輸入的支出項目:')
            while types!='0' :
                if types.isdecimal():#判斷是否輸入格式有誤
                    typesint=int(types)
                    if typesint==1 or typesint==2 or typesint==3 or typesint==4:#判斷是否輸入格式有誤
                        out=input("輸入支出")
                        if out.isdecimal():
                            outint=int(out)
                            if typesint==1:#將第一項支出放入對應串列
                                list1.append(outint)
                                count=count+1
                            elif typesint==2:#將第二項支出放入對應串列
                                list2.append(outint)
                                count=count+1
                            elif typesint==3:#將第三項支出放入對應串列
                                list3.append(outint)
                                count=count+1
                            elif typesint==4:#將第四項支出放入對應串列
                                list4.append(outint)
                                count=count+1
                        else:#支出輸入錯誤重新輸入
                            out=input("輸入支出")
                    else:#輸入的型別代號錯誤，重新輸入
                        print('1.飲食 2.日用品 3.娛樂 4.其他 0:離開')
                        types=input('請選擇要輸入的支出項目:')
                else:#輸入的型別代號錯誤，重新輸入
                    print('1.飲食 2.日用品 3.娛樂 4.其他 0:離開')
                    types=input('請選擇要輸入的支出項目:')
                print('1.飲食 2.日用品 3.娛樂 4.其他 0:離開')
                types=input('請選擇要輸入的支出項目:')#使迴圈可以繼續進行
            if count!=0:#計數器不等於0，有輸入支出
                a1=sum(list1)#第一項支出的總和
                a2=sum(list2)#第二項支出的總和
                a3=sum(list3)#第三項支出的總和
                a4=sum(list4)#第四項支出的總和
                All=a1+a2+a3+a4#總支出
                if a1>yy1:#判斷是否超出每項預算
                    print("花費超出預算"+str(a1-yy1)+"元")
                else:
                    print("花費小於預算"+str(yy1-a1)+"元")

                if a2>yy2:
                    print("花費超出預算"+str(a2-yy2)+"元")
                else:
                    print("花費小於預算"+str(yy2-a2)+"元")       

                if a3>yy3:
                    print("花費超出娛樂預算"+str(a3-yy3)+"元")
                else:
                    print("花費小於娛樂預算"+str(yy3-a3)+"元")

                if a4>yy4:
                    print("花費超出其他預算"+str(a4-yy4)+"元")
                else:
                    print("花費小於其他預算"+str(yy4-a4)+"元")
                #印製各項支出比例分析
                print('{0:<s}{1:^10s}{2:>s}' .format('支出項目名稱','支出','比例'))  #支出項目名稱名稱靠右對齊，支出置中對齊，比例向右對齊
                print('-'*30)
                print('{0:<s}{1:^10d}{2:>d}' .format('飲食',a1,ratio(a1,All)),'%')
                print('{0:<s}{1:^10d}{2:>d}' .format('日用品',a2,ratio(a2,All)),'%')
                print('{0:<s}{1:^10d}{2:>d}' .format('娛樂',a3,ratio(a3,All)),'%')
                print('{0:<s}{1:^10d}{2:>d}' .format('其他',a4,ratio(a4,All)),'%')
                mode=input('請輸入想要的功能(輸入代碼1或2)(要離開請輸入0)')
                if mode=='0':
                    break

                
        else:
            mode=input('請輸入想要的功能(輸入代碼1或2)(要離開請輸入0)')
    else:#功能代號輸入錯誤
        mode=input('請輸入想要的功能(輸入代碼1或2)(要離開請輸入0)')
```
