題目:1A2B猜數字遊戲

請先由出題者input一個每位數字皆由0-9組成且互不相同的四位數(ex:0184,1783,四個數字須不相同,0開頭沒關係)。

並設方法檢測試輸入格式有沒有錯，若輸入格式不符合則print("您輸入的格式有誤")，並重複執行讓出題者input直到格式無誤。

若無誤則打30個換行，讓遊戲者沒辦法直接看到題目，然後再讓遊戲者input。

接下來遊戲者一樣input一個每位數字皆由0-9組成且互不相同的四位數(和出題者一樣格式的)。

一樣檢驗輸入格式有沒有錯，若有錯則print("您輸入的格式有誤")，若無誤遊戲開始。

判斷遊戲者輸入的四個數字分別和出題者的四個數字是否數字相同且位置相同。

若數字相同且位置相同就是A。

若數字相同但是位置不同就是B。

若數字不同就什麼都沒有 以此判斷遊戲者是"幾A幾B"，並印出來。

例如: 出題者題目1920，遊戲者回答1234，就print("1A1B")，因為1數字相同且位置相同是一個A，2數字相同但位置不同是一個B，3和4就什麼都沒有。

遊戲者回答5678，就print("0A0B")。

遊戲者回答1290，就print("2A2B")。

遊戲者回答1209，就print("2A2B")。

直到遊戲者回答1920答對，就print("恭喜您答對了!")，遊戲結束。

重複執行讓遊戲者input，直到猜出正確答案為止。

```python
X='a'
Y='S'
while (X!="A"):
    X=(input("請出題者輸入一組每位數字皆由0-9組成且互不相同的四位數:"))
    if X.isdecimal():
        print( "-------------------------------------\n"
               "-------------------------------------\n"
               "-------------------------------------\n"
               "-------------------------------------\n"
               "-------------------------------------\n"
               "-------------------------------------\n"
               "-------------------------------------\n"
               "-------------------------------------\n"
              )
        break
    else:
         print("輸入格式錯誤，請再輸入一次")
while 2:
    Y=(input("請遊戲者輸入一組每位數字皆由0-9組成且互不相同的四位數:"))
    if X == Y:
        print("牛逼!4A0B，恭喜你答對了!")
        break
    else:
        A=0
        B=0
        for i in range(4):        #算出相同位置數字相等的數字有幾個
            if (str(X)[i] == Y[i]):
                A += 1
        for j in range(4): #算出數字相等的有幾個
            for k in range(4):
                if (str(X)[j] == Y[k]):
                    B += 1
                    
        B = B - A# B 為數字相等位置不同的數量
        print("%dA%dB" % (A, B),"再猜一次")
```
